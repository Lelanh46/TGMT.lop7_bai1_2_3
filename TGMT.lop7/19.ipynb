```python
def display(title, img):
    cv2.imshow(title, img)
    # Ch mt khong thi gian
    cv2.waitKey(0)
    # ng window
    cv2.destroyWindow(title)
```

```python
# Import th vin
import cv2
```

```python
# c mt nh (nhan TAB de nhin autocomplete)
path= 'C:/Users/ADMIN/TGMT.lop7/anh/1.jpg'
img = cv2.imread(path)
# In thng tin ca nh
print(img)
```

```python
# Ly kch thc
print(img.shape)
x, y, z = img.shape
print(x)
print(y)
print(z)
```

```python
# Hin th nh
cv2.imshow('Window', img)
# Ch mt khong thi gian
cv2.waitKey(10000)
# ng window
cv2.destroyWindow('Window')
```

```python
# Tch mu
b, g, r = cv2.split(img)
cv2.imshow('Window', img)
cv2.imshow('Red', r)
cv2.imshow('Green', g)
cv2.imshow('Blue', b)

# Ch mt khong thi gian
cv2.waitKey(0)
# ng window
cv2.destroyAllWindows()
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2.imshow('Window', img)
cv2.imshow('gray_img', gray_img)
cv2.imshow('img_2', img_2)
# Ch mt khong thi gian
cv2.waitKey(0)
# ng window
cv2.destroyAllWindows()
```

```python
# Thay i thng s ca im nh
height, width, z = img.shape
quater_h = height/2
quater_w = width/2

# geen color
green_color = (0,255,0)

# copy nh
img2 = img.copy()

# thay i mu ca gc 1/4 bn tri trn

for y in range(int(quater_h)):
    for x in range(int(quater_w)):
        img2[y, x] = green_color
# Hin th
display('Window', img)
display('Window-2', img2)
```
